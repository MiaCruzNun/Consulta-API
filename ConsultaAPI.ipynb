```python
url="https://api.datos.gob.mx/v1/inai.solicitudes"
r= requests.get(url)
response = r.text
print(response)
```

```python
j = r.json()
print(j)



```

```python
import pandas as pd

new = pd.json_normalize(j["results"])

print(new.head())


```

```python
print(new.columns)
```

```python
new.groupby(["DEPENDENCIA","ESTADO"]).nunique().sort_values("_id",ascending=False)
```

```python
new.groupby(["DEPENDENCIA","PAIS"]).count().sort_values("_id",ascending=False)
```

```python
from datetime import datetime

for i in range(0 , len(new.FECHARESPUESTA)):
    new["FECHASOLICITUD_DD/MM/YYYY"][i] = datetime.strptime(new.FECHASOLICITUD[i],"%d/%m/%Y %H:%M").date()
```

```python
type(new["FECHASOLICITUD_DD/MM/YYYY"])
```

```python
type(new.FECHARESPUESTA)
```

```python
new.FECHARESPUESTA -new["FECHASOLICITUD_DD/MM/YYYY"]
```

```python
new["TiempoRespuesta"] =(new.FECHARESPUESTA -new["FECHASOLICITUD_DD/MM/YYYY"])
```

```python
new.TiempoRespuesta.describe()
```

```python
new.groupby('PAIS')['TiempoRespuesta'].describe()
```

```python
new.groupby('ESTADO')['TiempoRespuesta'].describe()
```

```python
new.groupby('MEDIOENTREGA')['TiempoRespuesta'].describe().sort_values("mean",ascending=False)
```
